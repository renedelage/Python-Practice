```python
from datetime import date
import random

# define user and admin databases
user_db = {
    'user1' : 'Welcome123'
}
admin_db = {
    'admin' : 'Admin123'
}

# user login function
def user_login():
    useraccess = False
    count = 0
    while useraccess == False and count < 4:
        un = input("Username: ")
        pw = input("Password: ")
        if un in user_db and user_db[un] == pw:
            useraccess = True
            count = 4
            print("Login successful. User access granted.")
            user_shopping()
        else:
            count += 1
            print(f'Incorrect username or password entered. {4 - count} attempts remaining.')       
    if useraccess == False:
        print("Login failed. Goodbye.")

# define admin function + verify and grant admin access
adminaccess = False
def admin_login():
    global adminaccess
    adminaccess = False
    count = 0
    while adminaccess == False and count < 4:
        una = input("Username: ")
        pwa = input("Password: ")
        if una in admin_db and admin_db[una] == pwa:
            adminaccess = True
            count = 4
            print("Login successful.")
        else:
            count += 1
            print(f'Incorrect username or password entered. {4 - count} attempts remaining.')
    if adminaccess == True:
        print("Admin Access granted.")
        admin_portal()
    else:
        print("Login failed. Goodbye.")
            
# define session generator function
session_list = []
def generate_session_id():
    global session_list
    session = random.randint(1, 100000)
    if session not in session_list:
        session_list.append(session)
    else:
        while session in session_list:
            session = random.randint(1, 100000)
        session_list.append(session)
    return session

# create function to allow for user or admin login
def logon():
    loginauth = 0
    print("\nWelcome to Demo Marketplace")
    while loginauth == 0:
        login = input("\n'1' to Login as User. '2' to Login as Admin. '0' to Exit.\nPlease make a selection: ")
        if login == '0':
            print("Thank you for stopping by. Have a great day!")
            print("Goodbye.")
            loginauth = 1
        elif login == '1':
            loginauth = 1
            user_login()
        elif login == '2':
            loginauth = 1
            admin_login()
        else:
            print("Error. Please try again.")
    
# create product catalog
catalog = [
    {'id' : '1', 'name' : 'Boots', 'category' : 'Clothing', 'price' : 80},
    {'id' : '2', 'name' : 'Coat', 'category' : 'Clothing', 'price' : 180},
    {'id' : '3', 'name' : 'Jacket', 'category' : 'Clothing', 'price' : 125},
    {'id' : '4', 'name' : 'Cap', 'category' : 'Clothing', 'price' : 30},
    {'id' : '5', 'name' : 'Vintage T-Shirt', 'category' : 'Clothing', 'price' : 40},
    {'id' : '6', 'name' : 'Sunglasses', 'category' : 'Accessories', 'price' : 85}
]

catalog_id = {'1', '2', '3', '4', '5', '6'}

def update_catalog():
    global catalog
    for x in catalog:
        catalog_id.add(x['id'])

# create a function to display the catalog
def display_catalog():
    for x in catalog:
        print(f'Item ID #: {x['id']}, Product: {x['name']} ({x['category']}), Price: ${x['price']}')

cart = []

def display_cart():
    item = 0
    total = 0
    for x in cart:
        item += 1
        print(f'Item {item}: {x['name']}, Price: ${x['price']} ({x['category']})')
        total += x['price']
    print(f'Total: ${total}')

# create functions to manage cart
def add_to_cart():
    global cart
    shop = True
    update_catalog()
    while shop == True:
        selection = input("\nEnter the Item ID # of the item you'd like to purchase ('0' to exit): ")
        if selection == '0':
            print("Thank you for shopping at the Demo Marketplace!")
            shop = False
        elif selection in catalog_id:
            for x in catalog:
                if x['id'] == selection:
                    cart.append(x)
                    print(f'{x['name']} has been added to your cart.')
        else:
            print('Invalid selection. Please select a valid item.')

def delete_from_cart():
    global cart
    delete = True
    while delete == True:
        selection = input("\nEnter the Item ID # of the item you'd like to remove ('0' to exit): ")
        if selection == '0':
            print("Thank you for shopping at the Demo Marketplace!")
            delete = False
        elif selection in catalog_id:
            for x in cart:
                if x['id'] == selection:
                    cart.remove(x)
                    print(f'{x['name']} has been removed from your cart.')
                    break
        else:
            print('Invalid selection. Please select a valid item.')

def receipt():
    item = 0
    total = 0
    today = date.today()
    print(f'\nTransaction: {generate_session_id()}')
    print(f'Date: {today}\n ')
    for x in cart:
        item += 1
        print(f'Item {item}: {x['name']}, Price: ${x['price']}')
        total += x['price']
    print(f'Total: ${total}')
    print(f'Amount Paid: ${total}')
    print("Thank you!!")

# define the checkout function
def checkout():
    global cart
    payment = False
    display_cart()
    while payment == False:
        selection = input("\n'1' to Pay with Credit Card. '2' to Pay with Cash. '3' to Cancel and Return to Store. '0' to Leave.\nPlease make a selection: ")
        if selection == '0':
            print("Thank you for stopping at Demo Marketplace. Have a great day!")
            payment = True
        elif selection == '1':
            pin = input("Please enter your pin: ")
            print("Transaction successful. Thank you for shopping at Demo Marketplace. Have a great day!")
            receipt()
            cart.clear()
            payment = True
        elif selection == '2':
            print("Thank you for shopping at Demo Marketplace. Have a great day!")
            receipt()
            cart.clear()
            payment = True
        elif selection == '3':
            payment == True
            user_shopping()
        else: 
            print("Invalid selection. Please try again.")
    
# create function for user to interact with the store
def user_shopping():
    global cart
    shopping = True
    update_catalog()
    while shopping == True:
        selection = input("\n'1' to View Catalog. '2' to View Cart. '3' to Add to Cart. '4' to Remove from Cart. '5' to Checkout. '0' to Exit. \nPlease make a selection: ")
        if selection == '0':
            print("Thank you for shopping at Demo Marketplace!")
            cart.clear()
            shopping = False
            logon()
        elif selection == '1':
            display_catalog()
        elif selection == '2':
            display_cart()
        elif selection == '3':
            add_to_cart()
        elif selection == '4':
            delete_from_cart()
        elif selection == '5':
            checkout()
        else:
            print("Invalid selection. Please try again.")

# create a function for various admin functions (and defined those functions)
def admin_portal():
    global adminaccess
    manage = True
    while manage == True and adminaccess == True:
        selection = input("\n'1' to Add Items to Catalog. '2' to Remove Items from Catalog. '3' to Edit Items in Catalog. '4' to View Catalog. '0' to Logout. \nPlease make a selection: ")
        if selection == '0':
            manage = False
            adminaccess = False
            loginauth = 0
            logon()
        elif selection == '1':
            add_to_catalog()
        elif selection == '2':
            remove_from_catalog()
        elif selection == '3':
            edit_catalog()
        elif selection == '4':
            display_catalog()
        else:
            print("Invalid selection. Please try again.")

def add_to_catalog():
    global catalog
    adding = True
    while adding == True:
        selection = input("\n'1' to Add an Item. '2' to View Catalog. '0' to Return to Main Menu.\nPlease make a selection: ")
        if selection == '0':
            print("All additions complete.")
            adding = False
            admin_portal()
        elif selection == '1':
            pid = input("Enter the product's ID number: ")
            while pid in catalog_id:
                pid = input("That ID Number is already assigned to a Product. Please enter a new Number: ")
            pnm = input("Enter the product's name: ")
            pct = input("Enter the product's category: ")
            prc = int(input("Enter the product's price: "))
            new_prod = {'id' : pid, 'name' : pnm, 'category' : pct, 'price' : prc}
            catalog.append(new_prod)
        elif selection == '2':
            display_catalog()
        else:
            print("Invalid selection. Please try again.") 
        update_catalog()

def remove_from_catalog():
    global catalog
    removing = True
    while removing == True:
        selection = input("\n'1' to Remove and Item from Catalog. '2' to View Catalog. '0' to Return to Main Menu.\nPlease make a selection: ")
        if selection == '0':
            print("All removals complete.")
            removing = False
            admin_portal()
        elif selection == '1':
            rmv = input("Enter the ID Number of the product you wish to remove: ")
            if rmv in catalog_id:
                for x in catalog:
                    if x['id'] == rmv:
                        print(f'{x['name']} has been removed from the Catalog.')
                        catalog.remove(x)
                        catalog_id.remove(x['id'])
            else:
                print("No matching Product found.")
        elif selection == '2':
            display_catalog()
        else:
            print("Invalid selection. Please try again.")   

def edit_catalog():
    global catalog
    editing = True
    while editing == True:
        selection = input("\n'1' to Edit a Product's ID Number. '2' to Edit a Product's Name.\n'3' to Edit a Product's Category. '4' to Edit a Product's Price.\n'5' to View Catalog. '0' to Return to Main Menu.\nPlease make a selection: ")
        if selection == '0':
            print("Editing complete.")
            editing = False
            admin_portal()
        
        elif selection == '1':
            pid =  input("What is the current ID Number of the Product? ")
            if pid in catalog_id:
                for x in catalog:
                    if x['id'] == pid:
                        catalog_id.remove(pid)
                        nid = input(f'Please enter the new ID Number for {x['name']}: ')
                        if nid not in catalog_id:
                            x['id'] = nid
                            print("Reassignment complete.")
                            update_catalog()
                        else:
                            while nid in catalog_id:
                                nid = input("Error. Please choose a NUMBER not assigned to another Product: ")
                            x['id'] = nid
                            print("Reassignment complete.")
                            update_catalog()
            else:
                print("Error. Product not found.")
                                 
        elif selection == '2':
            pid = input("What is the ID Number of the Product you want to rename? ")
            if pid in catalog_id:
                for x in catalog:
                    if x['id'] == pid:
                        x['name'] = input(f'Assign a new Name for {x['name']}: ')
                        print("Reassignment complete.")
            else:
                print("Error. Product not found.")
                    
        elif selection == '3':
            pid = input("What is the ID Number of the Product you want to recategorize? ")
            if pid in catalog_id:
                for x in catalog:
                    if x['id'] == pid:
                        x['category'] = input(f'Assign a new Category for {x['name']}: ')
                        print("Reassignment complete.")
            else:
                print("Error. Product not found.")
                
        elif selection == '4':
            pid = input("What is the ID Number of the Product you want to reprice? ")
            if pid in catalog_id:
                for x in catalog:
                    if x['id'] == pid:
                        x['price'] = int(input(f'Assign a new Price for {x['name']}: $'))
                        print("Reassignment complete.")
            else:
                print("Error. Product not found.")
                
        elif selection == '5':
            display_catalog()
            
        else:
            print("Invalid selection. Please try again.")
            
```
